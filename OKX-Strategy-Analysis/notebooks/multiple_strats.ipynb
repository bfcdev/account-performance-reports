# Example: Multiple Strategies - Single Simulation
Create a vbt portfolio sim with more than one backtest strategy so we can see how they interact together.

```python
from vectorbtpro import *
import pandas as pd
```

```python
pf1 = vbt.load('/Users/ericervin/Documents/Coding/account-performance-reports/OKX-Strategy-Analysis/results/All-Baker-Duck-orders_BTCUSDT_pf.pkl')
pf2 = vbt.load('/Users/ericervin/Documents/Coding/account-performance-reports/OKX-Strategy-Analysis/results/Average-Moon-Cypress-orders_BTCUSDT_pf.pkl')
```

```python
pf1.stats()
```

```python
pf2.stats()
```

```python
strategy_keys = pd.Index(["All-Baker-Duck", "Avg-Moon-Cyprus"], name="strategy")

stacked_pf = vbt.PF.column_stack((pf1, pf2), wrapper_kwargs=dict(keys=strategy_keys))

grouped_pf = vbt.PF.column_stack(
    (pf1, pf2), 
    wrapper_kwargs=dict(keys=strategy_keys), 
    group_by=strategy_keys.name  
)

cash_sharing_pf = vbt.PF.column_stack(
    (pf1, pf2), 
    wrapper_kwargs=dict(keys=strategy_keys), 
    cash_sharing=True
)
```

```python
stacked_pf.stats(agg_func=None)
```

```python
grouped_pf.stats(agg_func=None)
```

```python
stacked_pf.stats()
```
